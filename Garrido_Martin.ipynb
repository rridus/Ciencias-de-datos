```python
import numpy as np

# 1. Definición de la función (según tu repositorio GitHub: rridus/Ciencias-de-datos)
def la_moda(datos):
    """
    Estima la moda de una lista de datos.
    Extraído de: https://github.com/rridus/Ciencias-de-datos
    """
    frecuencias = {}
    for elemento in datos:
        if elemento in frecuencias:
            frecuencias[elemento] += 1
        else:
            frecuencias[elemento] = 1

    max_frecuencia = max(frecuencias.values())
    modas = [k for k, v in frecuencias.items() if v == max_frecuencia]

    return modas[0] if len(modas) == 1 else modas

# 2. Carga de datos del Cúmulo de Virgo
# El archivo contiene columnas como Name, Blue Mag, Type, etc. [cite: 22]
try:
    # Cargamos el archivo usando genfromtxt para manejar los strings de la columna 'Type'
    data = np.genfromtxt('cumulo_de_Virgo.dat', dtype=None, names=True, encoding='utf-8')

    # Extraer la columna 'Type' (tipo de galaxia según secuencia de Hubble) [cite: 26]
    tipos_galaxias = data['Type']

    # 3. Estimar la moda usando tu función personal [cite: 13, 31]
    moda_resultado = la_moda(tipos_galaxias)

    # 4. Imprimir el mensaje solicitado
    print(f"La moda del tipo de galaxia en el cúmulo de Virgo es: {moda_resultado}")

except FileNotFoundError:
    print("Error: Asegúrate de que el archivo 'cumulo_de_Virgo.dat' esté en la misma carpeta.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Cargar el catálogo de estrellas brillantes
# Columnas: HR, HD, Vmag, B-V, SpType
data_stars = np.genfromtxt('bsc_sel.dat', names=True, dtype=None, encoding='utf-8')

# 2. Definir los tipos espectrales en el orden solicitado
tipos_objetivo = ['A0V', 'A1V', 'A2V', 'G8III', 'K0III', 'K3III', 'K5III'] #

# 3. Filtrar los datos para cada tipo espectral
datos_por_tipo = []
for tipo in tipos_objetivo:
    # Seleccionamos el color B-V para las estrellas que coincidan con el tipo espectral
    # Nota: B-V suele tener un guión en el nombre del campo, verifica el header del archivo
    mask = (data_stars['SpType'] == tipo)
    seleccion = data_stars['BV'][mask] # Ajustar nombre de columna según el archivo
    datos_por_tipo.append(seleccion)

# 4. Creación del gráfico orientado a objetos
fig, ax = plt.subplots(figsize=(12, 7))

# Configuración de los valores atípicos (fliers)
# Use cruces de color verde para los outliers
flier_props = dict(marker='x', markeredgecolor='green', markersize=8)

# Crear el boxplot
ax.boxplot(datos_por_tipo,
           labels=tipos_objetivo, #
           flierprops=flier_props) #

# 5. Formato y etiquetas
ax.set_title('Distribución del Color B-V por Tipo Espectral', fontsize=14)
ax.set_xlabel('Tipo Espectral', fontsize=12)
ax.set_ylabel('Índice de Color B-V', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Asegúrate de que el Notebook se entregue ejecutado para ver el output
plt.tight_layout()
plt.show()
```
